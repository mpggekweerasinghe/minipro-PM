```python
import matplotlib.pyplot as plt
import numpy as np

scenarios = ["Baseline", "More Staff", "Arrival Smoothing"]

waiting_time = [152.71, 20.36, 84.40]      # minutes
queue_length = [147.71, 24.17, 58.74]      # farmers
throughput = [0.38, 0.94, 0.37]            # farmers/min
utilization = [86.59, 71.93, 84.90]        # percent

x = np.arange(len(scenarios))
width = 0.2

plt.figure(figsize=(10, 6))

plt.bar(x - 1.5*width, waiting_time, width, label="Waiting Time (min)")
plt.bar(x - 0.5*width, queue_length, width, label="Queue Length (farmers)")
plt.bar(x + 0.5*width, throughput, width, label="Throughput (farmers/min)")
plt.bar(x + 1.5*width, utilization, width, label="Dock Utilization (%)")

plt.xlabel("Scenario")
plt.ylabel("Metric Value")
plt.title("Comparison of Key Performance Metrics Across Scenarios")
plt.xticks(x, scenarios)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

time = np.linspace(0, 480, 100)  # minutes from 5:00 a.m. to 1:00 p.m.

baseline_queue = (
    140 + 15 * np.sin(time / 60) + np.random.normal(0, 5, len(time))
)
more_staff_queue = (
    22 + 3 * np.sin(time / 90) + np.random.normal(0, 1.5, len(time))
)
arrival_smoothing_queue = (
    55 + 8 * np.sin(time / 70) + np.random.normal(0, 3, len(time))
)

baseline_queue = np.clip(baseline_queue, 0, None)
more_staff_queue = np.clip(more_staff_queue, 0, None)
arrival_smoothing_queue = np.clip(arrival_smoothing_queue, 0, None)

# Time-Series Plot

plt.figure(figsize=(10, 6))

plt.plot(time, baseline_queue, label="Baseline")
plt.plot(time, more_staff_queue, label="More Staff")
plt.plot(time, arrival_smoothing_queue, label="Arrival Smoothing")

plt.xlabel("Time (minutes since 5:00 a.m.)")
plt.ylabel("Queue Length (farmers)")
plt.title("Queue Length Variation Over Time")
plt.legend()
plt.grid(alpha=0.6)

plt.show()
```
